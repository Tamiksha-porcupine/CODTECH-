```python
from sqlalchemy import create_engine
import pandas as pd

# ‚úÖ Step 1: Define your MySQL connection string
# Replace 'root', 'your_password', and 'your_database' as needed
username = 'root'
password = 'tamiksha'
host = 'localhost'
database = 'sets'

# ‚úÖ Step 2: Create SQLAlchemy engine
engine = create_engine(f"mysql+mysqlconnector://{username}:{password}@{host}/{database}")

# ‚úÖ Step 3: Read from MySQL into Pandas DataFrames
try:
    imdb_df = pd.read_sql("SELECT * FROM imdb", con=engine)
    genres_df = pd.read_sql("SELECT * FROM genres", con=engine)

    print("‚úÖ Data Loaded Successfully")

    print("\nüé¨ IMDB Table:")
    display(imdb_df.head())

    print("\nüéÆ GENRES Table:")
    display(genres_df.head())

except Exception as e:
    print("‚ùå Error occurred while fetching data:", e)

# ‚úÖ Step 4: Optional ‚Äî write data back to MySQL
# Example: Backup imdb table into a new one called imdb_backup
try:
    imdb_df.to_sql("imdb_backup", con=engine, if_exists="replace", index=False)
    print("‚úÖ imdb_backup table created in MySQL.")
except Exception as e:
    print("‚ùå Error writing to MySQL:", e)

# ‚úÖ Step 5: Close connection ‚Äî optional, handled by SQLAlchemy automatically
```
